Association Rule Mining

Association rule mining adalah prosedur yang dimaksudkan untuk menemukan pola yang sering, korelasi, asosiasi, atau struktur kausal dari kumpulan data yang ditemukan di berbagai jenis basis data seperti basis data relasional, basis data transaksional, dan bentuk penyimpanan data lainnya. Melihat serangkaian transaksi yang terjadi, association rule mining bertujuan untuk mengetahui sebuah pola item tertentu, yang berkaitan dengan transaksi lainya.

A.	Definisi Association Rule Mining

Menurut (Techopedia, 2014), adalah proses penambangan data untuk menemukan aturan yang dapat mengatur asosiasi dan objek kausal antara set item. Jadi dalam transaksi tertentu dengan beberapa item, ia mencoba menemukan aturan yang mengatur bagaimana atau mengapa item tersebut sering dibeli bersama.

B.	Types Of Association Rules In Data Mining

Menurut (Rai, 2022), Biasanya ada empat jenis aturan asosiasi dalam penambangan data.

	Generalized association rule

	Interval information association rules

	Quantitatives association rules

	Eclat Algorithm

Algoritma Eclat adalah algoritma data mining yang digunakan untuk mencari frequent item. Seperti yang telah kita ketahui, beberapa algoritma association rule mining menggunakan format data horizontal dan beberapa di antaranya menggunakan format data vertikal untuk pembangkitan frequent itemset. Eclat tidak bisa menggunakan database horizontal. Jika ada database horizontal, maka kita perlu mengubahnya menjadi database vertikal.
